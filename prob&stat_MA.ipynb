```python
import pandas as pd
import numpy as np

# Generate 10 rows of sample price data
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=10, freq='D')
prices = np.round(100 + np.cumsum(np.random.randn(10)), 2)

# Create DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Price': prices
})

# Calculate moving average (window of 3)
df['MA'] = df['Price'].rolling(window=3).mean().round(2)

# Display the data
print(df)
```
